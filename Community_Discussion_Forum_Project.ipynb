```python
import os

project_folder = "Community_Discussion_Forum_Project"

os.makedirs(project_folder, exist_ok=True)

print("Folder Created:", project_folder)
```

```python
import pandas as pd

users = pd.DataFrame({
    "User_ID":[1,2,3,4,5],
    "Username":["Aman","Riya","Karan","Sneha","Priya"],
    "Role":["Member","Member","Moderator","Member","Admin"]
})

discussions = pd.DataFrame({
    "Discussion_ID":[101,102,103,104],
    "Title":[
        "Web Development",
        "AI Community",
        "Placement Prep",
        "Hackathon Group"
    ],
    "Posts":[40,60,25,50]
})

messages = pd.DataFrame({
    "Room":["General","AI","General","Hackathon","Placement"],
    "Messages":[120,90,140,75,100]
})

print(users.head())
print(discussions.head())
print(messages.head())
```

```python
import os

project_folder = "Community_Discussion_Forum_Project"

os.makedirs(project_folder, exist_ok=True)

users.to_csv(
    f"{project_folder}/users.csv",
    index=False
)

discussions.to_csv(
    f"{project_folder}/discussions.csv",
    index=False
)

messages.to_csv(
    f"{project_folder}/messages.csv",
    index=False
)

print("Step 2 files saved successfully")
```

```python
os.listdir(project_folder)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(7,5))

plt.bar(
    discussions["Title"],
    discussions["Posts"]
)

plt.title("Discussion Activity")

plt.xlabel("Discussion")

plt.ylabel("Total Posts")

plt.xticks(rotation=20)

plt.show()
```

```python
plt.figure(figsize=(7,5))

plt.plot(
    messages["Room"],
    messages["Messages"],
    marker="o"
)

plt.title("Chat Messages Activity")

plt.xlabel("Room")

plt.ylabel("Messages")

plt.show()
```

```python
users.to_csv(
    f"{project_folder}/users.csv",
    index=False
)

discussions.to_csv(
    f"{project_folder}/discussions.csv",
    index=False
)

messages.to_csv(
    f"{project_folder}/messages.csv",
    index=False
)

print("CSV Files Saved")
```

```python
plt.figure(figsize=(7,5))

plt.bar(
    discussions["Title"],
    discussions["Posts"]
)

plt.savefig(
    f"{project_folder}/discussion_graph.png"
)

print("Graph Saved")
```
